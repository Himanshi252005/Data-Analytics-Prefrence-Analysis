```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import pytz

```

```python
df = pd.read_csv("job_descriptions.csv")
```

```python
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
```

```python
df.describe()
```

```python
df.columns
```

```python
df['Salary_clean'] = (
    df['Salary Range']
    .astype(str)
    .str.replace(',', '', regex=True)
    .str.extract(r'(\d+)')
    .astype(float)
)
```

```python
df['Experience_clean'] = (
    df['Experience']
    .astype(str)
    .str.extract(r'(\d+)')   # extract first number
    .astype(float)
)
```

```python
filtered_df = df[
    (df['Work Type'] == 'Intern') &
    (df['latitude'] < 10) &
    (~df['Country'].str.startswith(('A','B','C','D'), na=False)) &
    (df['Company Size'] < 50000) &
    (df['Salary_clean'] > 9000) &
    (df['Experience_clean'] % 2 == 0) &
    (df['Job Posting Date'].dt.month.isin([1,3,5,7,9,11])) &
    (~df['Job Title'].str.contains(' ', na=False)) &
    (df['Job Title'].str.len() < 10)
]
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist).time()

start_time = pd.to_datetime("15:00").time()
end_time = pd.to_datetime("17:00").time()
```

```python
if start_time <= current_time <= end_time:
    preference_counts = (
        filtered_df['Preference']
        .value_counts()
        .sort_values(ascending=False)
    )

    plt.figure()
    preference_counts.plot(kind='bar')
    plt.xlabel('Preference')
    plt.ylabel('Count')
    plt.title('Preference vs Work Type (Intern)')
    plt.show()
else:
    print(" Chart is only visible between 3 PM and 5 PM IST.")
```
